```python
import pandas as pd
```

#1. Data Cleaning with Pandas

## Loading the csv dataset

```python
csv1 = pd.read_csv ('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
csv1
```

## Identifying and handling missing values in the dataset

```python
# check for missing values and count them
missing_values = csv1.isna().sum()

# calculate the percentage of missing values
percentage_missing = (missing_values / len(csv1)) * 100

# Display of the results
print("Missing Values:")
print(missing_values)
print("\nPercent of Missing Values:")
print(percentage_missing)
```

##Removing any duplicate rows and columns if they exist

```python
# Remove duplicate rows
csv1_no_duplicates = csv1.drop_duplicates()

# Check for duplicate columns
duplicate_columns = csv1_no_duplicates.columns[csv1_no_duplicates.columns.duplicated()]

# Remove duplicate columns
csv1_no_duplicate_columns = csv1_no_duplicates[csv1_no_duplicates.columns[~csv1_no_duplicates.columns.duplicated()]]
```

##Cleaning column names

```python
# Clean the column names
csv1.columns = csv1.columns.str.lower()
csv1.columns = csv1.columns.str.strip()
csv1.columns = csv1.columns.str.replace(' ', '_')
csv1.columns = csv1.columns.str.replace('[^a-zA-Z0-9_]', '', regex=True)

# Print the updated column names
print(csv1.columns)
```

#2. Data Transformation

##Creating new columns based on existing ones

```python
#Check the data type of the 'patient_age' column
print(csv1['patient_age'].dtype)

# Convert the 'patient_age' column to numeric
csv1['patient_age'] = pd.to_numeric(csv1['patient_age'], errors = 'coerce')

# Check the data type of the 'patient_age' column after conversion
print(csv1['patient_age'].dtype)

# Define the threshold for geriatric adults
threshold_65 = 65

# Create new column "Age_Group_Status" based on whether the patients are geriatric adults or not based on threshold
csv1['Age_Group_Status'] = csv1['patient_age'].apply(lambda x: 'Geriatric' if x >= threshold_65 else 'Not_Geriatric')

#Print the dataframe with the new 'Age_Group_Status' column
print(csv1)
```

##Aggregating data using groupby and compute summary statistics

```python
# Define the numeric columns for which to calculate the mean
numeric_columns = ['patient_age', 'average_heart_rate', 'average_bp', 'height_in_cm', 'weight_in_kg', 'visit_duration_mins', 'number_of_tests', 'prescription_cost_']

# Group the data by 'gender' and calculate the mean for the numeric columns
grouped_data = csv1.groupby('gender')
summary_stats = grouped_data[numeric_columns].mean()

# Reset the index
summary_stats.reset_index(inplace=True)

# Print the summary statistics
print(summary_stats)
```

##Cross-tabulation using multidimentional analysis

```python
# Create a cross-tabulation table for 'Gender' and 'Age_Group'
cross_tab = pd.crosstab(csv1['gender'], csv1['patient_age'])

# Display the cross-tabulation table
cross_tab
```

##Introduction to Alternative Libraries

```python
# Import polars
import polars as pl

#Import time
import time

# Define the URL of CSV dataset
csv1_url = 'https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv'

# Measure the start time
start_time = time.time()

# Load dataset using polars
csv1 = pl.read_csv(csv1_url)

# Measure the end time
end_time = time.time()

# Calculate the load time
load_time = end_time - start_time

# Display the start and end load times as well as the total load times
print(f"Start Time: {start_time}")
print(f"End Time: {end_time}")
print(f"Load Time: {load_time: .2f} seconds")

# Display the first few rows of the loaded dataset for verification
print("\nLoaded Dataset:")
print(csv1.head())
```
